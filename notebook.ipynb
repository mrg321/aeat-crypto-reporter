# Conexión de VS Code a Google Colab Cloud Runtime (sólo si lo quieres ejecutar desde vscode)

## 1. Instala las extensiones (sólo si lo quieres ejecutar desde vscode)

En Visual Studio Code:

1. Abre **Extensions**.
2. Busca **Google Colab**.
3. Instala la extensión oficial de Google.
4. Instala también **Jupyter** si VS Code lo solicita.

---

## 2. Abre tu proyecto en VS Code

Abre la carpeta donde tienes tu proyecto:

```text
File > Open Folder
```

Abre este notebook en vscode o súbelo a tu carpeta de Google Colab:

[Google Colab](https://colab.research.google.com/drive/)

```text
notebook.ipynb
```
---

## 3. Selecciona un kernel de Colab (sólo si lo quieres ejecutar desde vscode)

En el notebook abierto en VS Code:

Pulsa Select Kernel arriba a la derecha.
Elige Colab.
Inicia sesión con tu cuenta de Google.
Selecciona el runtime deseado:
CPU
GPU
TPU (si está disponible)

Desde ese momento, las celdas que ejecutes en VS Code se ejecutarán en el entorno cloud de Google Colab.
